# Car Price Analysis
We will be downloading the car price dataset, basically from the uci-automobile data archive url. Then we'll be taking a look into the data and cleaning the dataset to make it more understandable.

```python
# importing the required libraries for the purpose
import pandas as pd
import numpy as np
```

```python
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(path)
```

```python
df
```

The dataset is difficult to understand as it doesn't have headers. So we'll be adding headers manually so that the dataset makes more sense.  

This information is available at: https://archive.ics.uci.edu/ml/datasets/Automobile.

Hence, we'll use the headers from there and add them to the dataset

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df.columns = headers
df.head(10)
```

The ? symbol is used wherever the data is unavailable. Hence, we can replace it by using np.NaN to show that the data is not available.

```python
df = df.replace('?', np.NaN)
df.head(5)
```

```python
# Now we can drop the missing values rows using dropna
df = df.dropna(subset=['price'], axis=0)
df.head(10)
```

```python
# To get a basic understanding of the type of data we're working on we use info()
df.info()
```

Finally, we save the dataset into our machines to continue working on the dataset

```python
df.to_csv("car-data.csv")
```
